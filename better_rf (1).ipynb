```python
def read_csv(filename):
    return list(csv.reader(open(filename,newline='\n')))
def write_result(filename,result):
    with open(filename,'w',newline='') as csvfile:
        spamwriter = csv.writer(csvfile)
        spamwriter.writerow(['Id','Predicted'])
        for x in range(1,len(result)+1):
            spamwriter.writerow([x,result[x-1]])
def compress(list2):
    return np.array(list2)

def str_arr_2_float(data_arr):
    return list(map(lambda x:list(map(lambda y:float(y),x)),data_arr))
train_data = read_csv("./data/train_data.csv")
print(1)
train_data = str_arr_2_float(train_data)
print(2)
train_label = read_csv("./data/train_label.csv")
train_label = str_arr_2_float(train_label)
train_label = list(map(lambda x:x[0],train_label))
test_data = read_csv("./data/test_data.csv")
test_data = str_arr_2_float(test_data)
```

```python
clf = RandomForestRegressor(max_features= 5, random_state=10, oob_score=True, warm_start=True, min_samples_split=250, min_samples_leaf=50, n_estimators=1000, verbose = 2, n_jobs =30)# leaf=30
```

```python
print(len(train_data),len(train_label))
clf.fit(train_data, train_label)
```
